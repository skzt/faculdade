```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os

DATADIR = "C:/Users/skzt/Documents/TCC/fotos"
IMG_SIZE_X = 50
IMG_SIZE_Y = 15

def printArray(arr,x, y):
    for i in range(y):
        tmp = ''
        for j in range(x):
            tmp += str(arr[i][j])
            tmp += ' '
        print(tmp)

a = cv2.imread(os.path.join(DATADIR, "teste5"".png"), cv2.IMREAD_GRAYSCALE)
b = cv2.resize(a, (IMG_SIZE_X, IMG_SIZE_Y))
plt.imshow(b, cmap="gray")
plt.show()
# TODO: criar array de treino, separan
num1 = []
num2 = []
num3 = []
num4 = []

for i in range(IMG_SIZE_Y):
    tmp1 = []
    tmp2 = []
    tmp3 = []
    tmp4 = []
    for j in range(24, IMG_SIZE_X):
        if j <= 29:
            tmp1.append(b[i][j])
        elif j <= 35:
            tmp2.append(b[i][j])
        elif j <= 41:
            tmp3.append(b[i][j])
        elif j <= 47:
            tmp4.append(b[i][j])
        else:
            continue
    num1.append(tmp1)
    num2.append(tmp2)
    num3.append(tmp3)
    num4.append(tmp4)
    
num1 = np.array(num1)
num2 = np.array(num2)
num3 = np.array(num3)
num4 = np.array(num4)
```
